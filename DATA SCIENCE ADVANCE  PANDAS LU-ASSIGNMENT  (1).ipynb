# How to get the minimum, 25th percentile, median, 75th, and max of a numeric series?

```python
import numpy as np
import pandas as pd

num = np.random.RandomState(100)
X = pd.Series(num.normal(10,2,30))
print('Original Series:')
print(X)

res =np.percentile(X, q=[0,25,50,75,100])
print("\nMininimum, 25th percentile, median, 75th, and maximum of a given numeric Series")
print(res)
```

# How to find the positions of numbers that are multiples of 3 from a series?

```python
import pandas as pd
import numpy as np

#creating random Series
abc=pd.Series(np.random.randint(1,25,15))
print("Original Series:")
print(abc)

#Finding the position of the numbers divisible by 3
res=np.argwhere(abc % 3==0)
print("Position of numbers that are multiples of 3:\n")
print(res)
```

# How to stack two series vertically and horizontally ?

```python
#Creating the Series
S1 = pd.Series(["Triangle","Square","Circle","Cube","Sphere"])
print("S1")
print(S1)

S2 = pd.Series([3,4,1,6,5])
print("S2")
print(S2)

#Stacking the Sries Horizontally
df = pd.concat([S1,S2], axis = 1)
print("\n Series Stacked Horizontally")
print(df)
```

```python
#Staking Series Vertically
S1 = pd.Series(["Triangle","Square","Circle","Cube","Sphere"])
print("S1")
print(S1)

S2 = pd.Series([3,4,1,6,5])
print("S2")
print(S2)

#Stacking the Sries Vertically
df = pd.concat([S1,S2], axis = 0)
print("\n Series Stacked Vertically")
print(df)
```

# How to get the day of month, week number, day of year and day of week from a series of date strings?

```python
from dateutil.parser import parse
date_series=pd.Series(["14 Dec 2002",'11-07-2003',"2007/12/04","2008-10-08", "12/01/2013"])
print("Series:")
print(date_series)
print()
date_series = date_series.map(lambda x: parse(x))
print("Day of the Month:")
print(date_series.dt.day.tolist())
print()
print("Day of the Year:")
print(date_series.dt.dayofyear.tolist())
print()
print("Week Number:")
print(date_series.dt.weekofyear.tolist())
print()
print("Day of Week:")
print(date_series.dt.day_name())
```

# How to compute the euclidean distance between two series?

```python
# Euclidean distance 
#Is a straight line distance between two points.

#Creating Series
a = pd.Series([2,3,4,5,6,7])
b = pd.Series([8,10,12,14,16,20])
Euc_dist=(np.linalg.norm(a-b))
print("Series 1:")
print(a)
print()
print("Series 2:")
print(b)
print()
print("Euclidean Distance between two series is :", Euc_dist)
```

# ScatterPlot for comparing GDP against CO2 emissions (per capita)

```python
# select GDP Per capita emissions for the United States
indicator = 'GDP per capita \(constant 2005'
country = 'USA'

mcn1 = data['IndicatorName'].str.contains(indicator) 
mcn2 = data['CountryCode'].str.contains(country)

# stage is just those indicators matching the USA for country code and CO2 emissions over time.
gdp_co= data[mcn1 & mcn2]

```

```python
gdp_co.head()
```

```python
# select CO2 emissions for the United States
indicator = 'CO2 emissions \(metric'
country = 'USA'

mcn1 = data['IndicatorName'].str.contains(indicator) 
mcn2 = data['CountryCode'].str.contains(country)

# state is just those indicators matching the USA for country code and CO2 emissions over time.
state = data[mcn1 & mcn2]
```

```python
print("GDP Min Year = ", gdp_co['Year'].min(), "max: ", gdp_co['Year'].max())
print("CO2 Min Year = ", state['Year'].min(), "max: ", state['Year'].max())
```

```python
#We do the truncation to resize 
gdp_co_trunc = gdp_co[gdp_co['Year'] < 2012]
print(len(gdp_co_trunc))
print(len(state))
```

```python
#Plotting a scatter plot comparing the GDP against CO2 per capita
fig, axis = plt.subplots()
# Xticks, Xlabel, Ylabel


axis.set_title('CO2 Emissions vs. GDP \(per capita\)',fontsize=12)
axis.set_xlabel(gdp_co_trunc['IndicatorName'].iloc[0],fontsize=12)
axis.set_ylabel(state['IndicatorName'].iloc[0],fontsize=12)

X = gdp_co_trunc['Value']
Y = state['Value']

axis.scatter(X, Y)
plt.show()
```
